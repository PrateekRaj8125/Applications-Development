## Lists

### List  Creation

```python
emptyList=[]
list1=['one','two','three','four']
list2=[1,2,3,4]
list3=[[1,2],[3,4]]
list4=[2305608,'Aryan Ranjan',22,8.2]
print("",emptyList,"\n",list1,"\n",list2,"\n",list3,"\n",list4)
```

### User-Input List

```python
L=eval(input("Enter a List:"))
print("The elements of list entered:",L)
```

### List Indexing

```python
lst = [1 , 2 ,3 ,4 ,5]
print(lst[0])
print(lst[-4])
```

### List Slicing

```python
numbers = [10 , 20 , 30 ,40 , 50 , 60 ,70 ,80]
print(numbers[::-2])
print(numbers[-4:-7])
print(numbers[-7:-4])
print(numbers[::2])
print(numbers[4:7])
print(numbers[7:4])
```

### List Mathematical Methods

```python
L=[20,30,40,50]
print("Sum:",sum(L))
print("Length of List:",len(L))
print("Average:",sum(L)/len(L))
print(min(L))
print(max(L))
```

### List Methods

```python
fruits = ['orange','banana', 'apple','pear','kiwi','apple','banana']
print(fruits.count('orange'))
print(fruits.count('licchi'))
print(fruits.index('banana'))
print(fruits.index('banana',3))
fruits.reverse()
print(fruits)
fruits.append('grape')
print(fruits)
```
