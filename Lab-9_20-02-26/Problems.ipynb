5. Write a program to print all the common elements of 2 list.


```python
list1=eval(input("Enter first List:"))
list2=eval(input("Enter second List:"))
result=[item for item in list1 if item in list2]
print("Common items:",result)

result2=set(list1)&set(list2)
print("Common items(using set()):",result2)
del list1
del list2
del result
del result2
```

27. Write a Python program to replace last value of tuples in a list.  
    Sample list: [(10, 20, 40), (40, 50, 60), (70, 80, 90)]  
    Expected Output: [(10, 20, 100), (40, 50, 100), (70, 80, 100)]  


```python
data = [(10, 20, 40), (40, 50, 60), (70, 80, 90)]

result = [t[:-1] + (100,) for t in data]
print(result)
```

28. Write a Python program to replace last value of tuples in a list.  
    Sample data: [(), (), ('',), ('a', 'b'), ('a', 'b', 'c'), ('d')]  
    Expected output: [('',), ('a', 'b'), ('a', 'b', 'c'), 'd']  


```python
data = [(), (), ('',), ('a', 'b'), ('a', 'b', 'c'), ('d',)]

result = [t for t in data if t]
print(result)
```

29. Write a Python program to sort a tuple by its float element.  
    Sample data: [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]  
    Expected Output: [('item3', '24.5'), ('item2', '15.10'), ('item1', '12.20')]  


```python
data = [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]

sorted_data = sorted(data, key=lambda x: float(x[1]), reverse=True)
print(sorted_data)
```

30. Write a Python program to count the elements in a list until an element is a tuple.

```python
data = [10, 20, 30, (40, 50), 60]

count = 0
for item in data:
    if isinstance(item, tuple):
        break
    count += 1

print(count)
```
