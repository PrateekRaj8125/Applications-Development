1. Write a python program to count the total number of digits in a given number.

```python
data=int(input("Enter a no.:"))
num=data
dctr=0
while data>0:
    dctr+=1
    data//=10
print("OUTPUT:")
print("No. of digits in",num,"is",dctr)
```

2. Write a python program to input 10 numbers into a set and display only the unique elements.

```python
data=[]
for i in range(10):
    n=int(input(f"Enter {i+1} th no.:"))
    data.append(n)
data_set=set(data)
print("OUTPUT:")
print("Original Elements set:",data)
unique=set([x for x in data if data.count(x)==1])
print("Set of unique elements:",unique)
```
