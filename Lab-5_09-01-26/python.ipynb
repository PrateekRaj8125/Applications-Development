1. Write a python program to input two numbers and do all basic arithmetic operations on
them.

```python
a=float(input("Enter first number: "))
b=float(input("Enter second number: "))
print(f"Addition:", a+b)
print(f"Subtraction:", a-b)
print(f"Multiplication:", a*b)
print(f"Division:", a/b)
print(f"Modulus:", a%b)
print(f"Exponent:", a**b)
print(f"Floor Divison:", a//b)
```

2. Write a python program to input two number and swap their values without using any third
variable.


```python
a=input("Enter value for a: ")
b=input("Enter value for b: ")
print(f"Before swapping: a={a}, b={b}")
a,b = b,a
print(f"After swapping: a={a}, b={b}")
```


4. Write a python program to input the basic salary of a person and compute its TA (20% of
basic), DA (120% of basic), HRA (30% of basic), Gross (basic + ta + da + hra).


```python
basic = float(input("Enter basic salary: ₹"))
ta = 0.20 * basic
da = 1.20 * basic
hra = 0.30 * basic
gross = basic + ta + da + hra
print(f"TA: ₹{ta}\n DA: ₹{da}\n HRA: ₹{hra}\n Gross Salary: ₹{gross}")
```


11. Write a python program to input a the sides of a triangle and print its area.


```python
import math
a=float(input("Side a: "))
b=float(input("Side b: "))
c=float(input("Side c: "))
s=(a+b+c)/2
area = math.sqrt(s*(s-a)*(s-b)*(s-c))
print(f"Area of triangle: {area}")
```


12. Write a python program to compute SI and CI.


```python
p = float(input("Principal: ₹"))
r = float(input("Rate: %p.a."))
t = float(input("Time (in years): "))
si = (p * r * t) / 100
ci = p * (pow((1 + r / 100), t)) - p

print(f"Simple Interest: ₹{si}")
print(f"Compound Interest: ₹{ci}")
```
