1. Code the following scenario

    “Guess a number game” :

    - The computer picks a random number from 1 to 5, the player tries to guess.
    The player may have the flexibility to enter the number in an expression format.


```python
from random import randint

while True:
    num = randint(1,5)
    guess = eval(input('Enter your guess [Number 1 - 5] : '))
    if guess==num:
        print('You got it!')
    else:
        print("The number is : ",num)
    i=input("Do you want to continue [y/n] : ")
    if(i == 'n' or i=='N'):
        print("Thanks for Playing ...")
        break
```


2. Print the Grade

    Read 3 subject marks, calculate average mark and print the grade based on the
    following conditions

    avg >= 90 Grade O
    80 <=avg < 90 Grade E
    70 <=avg < 80 Grade A
    Otherwise Grade F



```python
m1=float(input("Enter marks of Sunject 1:"))
m2=float(input("Enter Marks of Subject 2:"))
m3=float(input("Enter marks of Subject 3:"))
avg=(m1+m2+m3)/3
if avg>=90:
    print("Grade 0")
    exit
elif avg>=80:
    print("Grade E")
    exit
elif avg>=70:
    print("Grade ")
    exit
else:
    print("Grade F")
```


3. Kids Multiplication Table
    Write a multiplication game program for kids. The program should give the player ten
    randomly generated multiplication questions to do. After each, the program should tell
    them whether they got it right or wrong and what the correct answer is.

    Sample Output:

    Question 1: 3 x 4 = 12  
    Right!  
    Question 2: 8 x 6 = 44  
    Wrong. The answer is 48.  
    ...  
    ...  
    Question 10: 7 x 7 = 49  
    Right.  

```python
from random import randint
for i in range(1,11):
    num1=randint(1,10)
    num2=randint(1,10)
    prod=num1*num2
    ch=int(input(f"Question {i}: {num1} x {num2}="))
    if(prod==ch):
        print("Question",i,":",num1,"x",num2,"=",ch,"\nRight!")
    else:
        print("Question",i,":",num1,"x",num2,"=",ch,"\nWrong. The answer is",prod)
```
